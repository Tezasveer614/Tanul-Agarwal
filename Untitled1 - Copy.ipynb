```python
x="Welcome to LPU"
lower=0
upper=0
Space=0
l=len(x)
for i in range(0,l):
  a=x[i]
  if(a.islower()):
    lower += 1
  elif(a.isupper()):
    upper += 1
  elif(a==" "):
    Space += 1
print("Lower characters =",lower)
print("Upper characters =",upper)
print("Spaces =",Space)
```

```python
a=int(input("a: "))
b=int(input("b: "))
m,n=0,0
if(a>b):
  m=a
  n=b
else:
  m=b
  n=a
d=m%n
while(d>0):
  m=n
  n=d
  d=m%n
print("GCD of two numbers is=",n)
```
