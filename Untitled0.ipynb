```python
for a in range(2,40000):
  b=0
  for x in range(1,a):
    if(a%x==0):
      b+=x
  if(a==b):
    print(a,"is perfect number")
```

```python
N=int(input('Enter number of overs completed: '))
R=float(input('Enter run rate of N overs: '))
S=int(input('Target score: '))
n=50-N
s=S-(R*N)
r=s/n
print("Run rate for remaining overs = {0:.2f}".format(r))
```

```python
n=9
for i in range(1,n+1):
  for j in range(1,i+1):
    print('*',end=" ")
  print()
for m in range(n-1,0,-1):
  for p in range(1,m+1):
    print("*",end=" ")
  print()
```
